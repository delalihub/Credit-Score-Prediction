```python
def delayed_payments(x):
    if x <=0:
        return 'Never_delayed'
    elif x<= 7:
        return '1-7times'
    elif x <=12:
        return "8_12times"
    elif x<=20:
        return '13_20times'
    elif x >20:
        return "20plus"
```

## X-y splitting

```python
y = data['Credit_Score']
X = data.drop('Credit_Score', axis=1)
data = pd.concat((X,y), axis=1)
```

## train-test-splitting

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1337)
```

## Get numericals & categoricals (train and test)

```python
X_train_cat = X_train.select_dtypes(object)
X_test_cat = X_test.select_dtypes(object)
```

```python
X_train_num  = X_train.select_dtypes(np.number)
X_test_num = X_test.select_dtypes(np.number)
X_train_num.head()
```

## Use the MinMaxScaler to fit X_train_num, transform both x_train_num and x_test_num

```python
from sklearn.preprocessing import MinMaxScaler

transformer = MinMaxScaler().fit(X_train_num)     #1. Use the MinMaxScaler to fit X_train_num and name the variable ttransformer

X_train_num_scaled = transformer.transform(X_train_num)   #2a. use the transformer to transform (X_train_num)
X_test_num_scaled = transformer.transform(X_test_num)     #2b.use the transformer to transform (X_test_num)


X_train_num_scaled = pd.DataFrame(X_train_num_scaled, columns=list(X_train_num.columns))
X_test_num_scaled = pd.DataFrame(X_test_num_scaled, columns=list(X_test_num.columns))
```

#### Onehot encode categorical columns: X_train_cat and X_test_cat. fit only on the train set but encode both train x-set and x-set

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(handle_unknown='error', drop='first').fit(X_train_cat) #1. Use onehot encoder to fit the X_train_cat
X_train_cat_encoded = encoder.transform(X_train_cat).toarray()   #2a. transform the X_train_cat with the encoder
X_test_cat_encoded = encoder.transform(X_test_cat).toarray()  #2b. transform the X_test_cat with the encoder
cols = encoder.get_feature_names_out(input_features=X_train_cat.columns)
cols1 = encoder.get_feature_names_out(input_features=X_test_cat.columns)

# Put the onehot encoded in a dataframe
X_test_cat_encoded = pd.DataFrame(X_test_cat_encoded, columns = cols1)
#display(onehot_X_test_cat_encoded)
X_train_cat_encoded = pd.DataFrame(X_train_cat_encoded, columns = cols)
#onehot_X_train_cat_encoded
```

### Concat X_train_cat with X_train_num. Do same for X_test_cat and X_test_num

```python
X_train_scaled = pd.concat((X_train_num_scaled, X_train_cat_encoded ), axis = 1)
display(X_train_scaled)
X_test_scaled = pd.concat((X_test_num_scaled, X_test_cat_encoded ), axis = 1)
X_test_scaled
```

### checking for imbalance in target column

```python
data['Credit_Score'].value_counts()
```

### Balance imbalanced target column using SMOTE

```python
#!pip install imblearn
from imblearn.over_sampling import SMOTE
```

```python
# by deafult, takes a 5-neighbour KNN to build a new point, I decided to experiment with 3 here
BalancerSMOTE = SMOTE(random_state = 100)
X_train_SMOTE,y_train_SMOTE = BalancerSMOTE.fit_resample(X_train_scaled,y_train)
```

```python
X_train_SMOTE
```

```python
y_train_SMOTE.value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix

clf = RandomForestClassifier(max_depth=5,
                             min_samples_split=20,
                             min_samples_leaf =20,
                             max_samples=0.2,
                             random_state = 42)
clf.fit(X_train_SMOTE, y_train_SMOTE)
print(clf.score(X_train_SMOTE, y_train_SMOTE))
print(clf.score(X_test_scaled, y_test))

y_pred = clf.predict(X_test_scaled)
display(y_test.value_counts())
display(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_pred)
ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_).plot()
plt.show()
```

## Put predictions in dataframe

```python
All_X = pd.concat((X_train_scaled,X_test_scaled), axis = 0)
```

```python
#Predictions
predictions = clf.predict(All_X)
display(confusion_matrix(y, predictions))
```

```python
data["Predictions"] = clf.predict(All_X)


## show confun matrix(cm)
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y, predictions)
ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_).plot()
plt.show()
```
